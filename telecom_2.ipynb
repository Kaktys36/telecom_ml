# Проект для "Ниединогоразрыва.ком"

**Описание проекта:** Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. 

**Цель проекта:** построить модель, которая по персональным данным, которые собрала команда оператора (информация об их тарифах и договорах) определит уйдёт ли пользователь в ближайшее время. 

**Условия проекта:**
- Использовать минимум 3 модели машинного обучения и выбрать лучшую из них
- Достичь метрики AUC-ROC >= 0.85

**Наметим план проекта:** 
- Загрузим неодходимые данные и библиотеки
- Проведём исследовательский анализ 
- Подготовим данные для обучения:
    - Подготовим признаки (удалим наименее значимые, изменим тип и значения для модели при необходимости)
    - Исследуем наличие/отсутсвие дисбаланса классов (при необходимости решим эту проблему)
    - Разделим данные на выборки
- Обучим модели и подберём оптимальные гиперпараметры
- Сравним результаты моделей и сделаем анализ лучшей модели (важность признаков, матрица ошибок)
- Проверим её   
- Напишем выводы по проекту и дадим рекомендации заказчику 


**Описание данных (Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах. Они находятся в папке `/datasets/`):**

Данные состоят из файлов, полученных из разных источников:

- `contract_new.csv` — информация о договоре (информация о договорах актуальна на 1 февраля 2020):
    - `BeginDate` - дата начала действия договора
    - `EndDate` - дата окончания действия договора
    - `Type` - тип оплаты услуг (за услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года):
         - `Month-to-month` - оплата каждый месяц
         - `Two Year` - оплата за год
         - `One Year` - оплата за 2 года
    - `PaperlessBilling` - наличие/отсутствие электронного чека
    - `PaymentMethod` - метод оплаты (кредитная карта, банковский перевод и т.д.)
    - `MonthlyCharges` - месячное пополнение баланса
    - `TotalCharges` - общее пополнение счёта
    
    
- `personal_new.csv` — персональные данные клиента:
    - `gender` - пол клиента
    - `SeniorCitizen` - возраст клиента (1-пожилой, 0-не пожилой)
    - `Partner` - наличие/отсутствие партнёра 
    - `Dependents` - наличие/отсутствие иждивенцев 
    
    
- `internet_new.csv` — информация об интернет-услугах:
    - `InternetService` - тип подключея к интернет-сети:
        - `DSL` - через телефонную линию
        - `Fider optic` - оптоволоконный кабель
    - `OnlineSecurity` - наличие/отсутствие услуги OnlineSecurity (блокировки небезопасных сайтов)
    - `OnlineBackup` - наличие/отсутствие услуги OnlineBackup (облачное хранилище файлов для резервного копирования данных)
    - `DeviceProtection` - наличие/отсутствие услуги DeviceProtection (антивирус)
    - `TechSupport` - наличие/отсутствие услуги TechSupport (выделенная линия технической поддержки)
    - `StreamingTV` - наличие/отсутствие услуги StreamingTV (стриминговое телевидение)
    - `StreamingMovies` - наличие/отсутствие услуги StreamingMovies (каталог фильмов)
    
   
- `phone_new.csv` — информация об услугах телефонии:
    - `MultipleLines` - наличие/отсутствие услуги MultipleLines (подключение телефонного аппарата к нескольким линиям одновременно)

Во всех файлах столбец `customerID` содержит код клиента.

## Исследовательский анализ данных

### Первичный осмотр файлов

Для начала объединим наши датасеты в один, чтобы исследовать признаки и их взаимосвязи целиком. Посмотрим с какого датасета начать объединение. Для этого создадим функцию `first_view` которая отобразит основые данные.

```python
def first_view(dataset):
    print('Столбцы датасета:', dataset.columns, '\n')
    print('Количество пропусков в столбцах:\n', dataset.isna().sum(), '\n')
    print('Размер датасета:', dataset.shape)
    display(dataset.sample(n=15))
```

```python
datasets_name = {'contract_df': contract_df, 'personal_df': personal_df, 'internet_df': internet_df, 'phone_df': phone_df}

for i in datasets_name:
    print('+++++++++++++++++++++++++++++++++++++++++++++++++++', '\n', 'Название датасета:', i, '\n+++++++++++++++++++++++++++++++++++++++++++++++++++')
    first_view(datasets_name[i])
```

Мы видим что размер `contract_df` и `personal_df` самый большой-7043 строка, а вот количество строк в других датафреймах отличается. При этом в данных нет ни одного пропуска, но они появятся при объединении. Меньшее количество строк наблюдается в датафреймах `phone_df` и `internet_df`. Это связано с тем, что не все пользователи подключили интернет и телефонную связь одновременно.

Теперь посмотрим значимые признаки по phik-корреляции (будем использовать её так как она отслеживает нелинейные связи в отличие от обычной). После использования модели отобразим значимость признаков с помощью features_importance, а сейчас нам нужно понять как подготовить признаки для модели (признаков сейчас достаточно много и мы можем сократить время обучения и повысить метрику если избавимся от неважных данных). Но перед тем как посмотреть корреляцию добавим нам таргетный столбец.

```python
def client_leave(date):
    if date == 'No':
        return 0
    else:
        return 1
```

```python
df['Leave'] = df['EndDate'].apply(client_leave)
```

```python
phik = df.phik_matrix()
phik['Leave'].sort_values(ascending=False)
```

Мы видим что на наш таргет влияет дата прихода клиента (даже скорее не она, а количество дней клиента в компании, позже добавим этот признак в отдельный столбец), сумма потраченная за период пользования услугами компании (тоже связана со временем в компании), почти со всеми дополнительными услугами кроме OnlineSecurity и TechSupport, а также наблюдается  корреляция с методом оплаты.

#### SplitEncoder

Создадим класс `SpliEncoder`, внутри которого будут 4 функции:
- `split` для разбивки указанного датасета 
- `num_cat_cols` для определения категориальных и численных столбцов
- `encoder` для кодирования выборок, полученных внутри класса
- `get_encoded_feature_names` для получения списка названий закодированных столбцов

Благодаря этому классу мы сможем обработать разные датасеты для обучения моделей, подготовив их внутри класса и пропустить шаги подготовки разных датасетов.

```python
RANDOM_STATE = 220523
```

```python
class SplitEncoder:
    #Конструктор. Получает датафрейм и определяет все данные на выходе наших функций
    def __init__(self, data):
        self.data = data
        self.features, self.target, self.features_train, self.features_test, self.target_train, self.target_test = self.split()
        self.numeric_cols, self.categorical_cols = self.num_cat_cols()
        self.ohe_scaler = self.encoder()
        self.encoded_feature_names = self.get_encoded_feature_names
    
    #Функция split возвращает выборки, признаки и таргет
    def split(self):
        features = self.data.drop('Leave', axis=1)
        target = self.data['Leave']

        features_train, features_test, target_train, target_test = train_test_split(features, target, 
                                                                                    test_size=0.25, 
                                                                                    stratify=target,
                                                                                    random_state=RANDOM_STATE) 

        print('Длина обучающей выборки сокращённого датасета:', len(features_train), 'или', 
              '{:.1%}'.format(len(features_train) / len(self.data)), '\n' 
              'Длина тестовой выборки сокращённого датасета:', len(features_test), 'или', 
              '{:.1%}'.format(len(features_test) / len(self.data)))

        return features, target, features_train, features_test, target_train, target_test

    #Функция num_cat_cols возвращает списки категориальных и численных столбцов по переменной features
    def num_cat_cols(self):
        numeric_cols = self.features.select_dtypes(exclude=['object']).columns.tolist()
        categorical_cols = self.features.select_dtypes(include=['object']).columns.tolist()

        return numeric_cols, categorical_cols

    #Функция encoder возвращает обученный модифицированный кодировщик ohe_scaler, который кодирует отдельно 
    #категориальные и численные столбцы, полученные с помощью num_cat_cols. ohe_scaler позже будет передаваться в 
    #пайплайн при обучении модели
    def encoder(self):
        ohe_scaler = ColumnTransformer(
            [("numeric", StandardScaler(), self.numeric_cols), 
             ("categorical", OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False), self.categorical_cols)])

        return ohe_scaler.fit(self.features_train, self.target_train)
    
    #Функция для получения списка названий закодированных столбцов потребуется после для выявления важности признаков
    #обращается к ohe_scaler используя атрибут transformers_ и get_feature_names для получения названий категориальных
    #столбцов и обращается к numeric_cols, полученных внутри класса
    def get_encoded_feature_names(self):
        encoded_categorical_cols = list(self.ohe_scaler.transformers_[1][1].get_feature_names(self.categorical_cols))
        encoded_feature_names = self.numeric_cols + encoded_categorical_cols
        
        return encoded_feature_names
```

#### model_info

Создадим функцию `model_info`, которая будет записывать результаты работы наших моделей и их параметры.

```python
results = pd.DataFrame(columns=['Датасет', 'Модель', 'Время обучения', 'Время предсказания', 'AUC-ROC', 'Precision',
                                'Recall', 'F1', 'Лучшие параметры модели'])
results
```

```python
#На вход модель получает название используемого датасета, название используемой модели, саму предобученную модель
#с подобранными подобранными гиперпараметрами, треннировочные признаки датасета, треннировочный таргет, количество 
#фолдов для кросс-валидации в подсчёте метрик
def model_info(dataset, name, model, features_train, target_train, cv_amount):
    
    #Создание нового списка
    information = []
    
    #Получение имени датасета
    information.append(dataset)
    
    #Получение названия используемой модели
    information.append(name)
    
    #Получение времени обучения лучшей модели
    start = time.time()    
    model.best_estimator_.fit(features_train, target_train)
    fit_time = time.time() - start
    information.append(fit_time)
    
    #Получение времени предсказания лучшей модели
    start = time.time()  
    predicted = model.best_estimator_.predict(features_train)
    predict_time = time.time() - start
    information.append(predict_time)
    
    #Подсчёт auc-roc метрики с кросс-валидацией
    cv_prob = cross_val_predict(model.best_estimator_, features_train, target_train, cv=cv_amount, method='predict_proba')
    auc = auc_roc(target_train, cv_prob[:, 1])
    information.append(auc)
    
    #Подсчёт precission метрики с кросс-валидацией
    precision = cross_val_score(model.best_estimator_, features_train, target_train, cv=cv_amount, scoring='precision').mean()
    information.append(precision)
    
    #Подсчёт recall метрики с кросс-валидацией
    recall = cross_val_score(model.best_estimator_, features_train, target_train, cv=cv_amount, scoring='recall').mean()
    information.append(recall)
    
    #Подсчёт f1 метрики с кросс-валидацией
    f1 = cross_val_score(model.best_estimator_, features_train, target_train, cv=cv_amount, scoring='f1').mean()
    information.append(f1)
    
    #Получение гиперпараметров лучшей модели
    best_params = model.best_params_
    information.append(best_params)
    
    #Добавление новой строки в таблице results с данными полученными в функции
    results.loc[len(results.index)] = information
    
    return results
```

### Матрица ошибок и подбор порога 

Выведем матрицу ошибок лучшей модели.

```python
pred = optuna_search_lgbm_df.predict(features_test)

cm = confusion_matrix(target_test, pred)

plt.figure(figsize=(16, 9))
sb.heatmap(cm, annot=True, fmt="d", cmap=sb.color_palette("Blues"), cbar=False)

plt.title('Матрица ошибок', size=20)
plt.xlabel('Предсказанные значения', size=18)
plt.ylabel('Истинные значения', size=18)
plt.show()
```

На матрице ошибок видно что в основном проблема предсказаний заключается в ложноположительных значениях. Для бизнеса это значит следующее-если данная модель войдёт в продакшн, то методы удержания клиента будут применяться к тем клиентам, которые на самом деле не собираются уходить. Попробуем подобрать порог классификации с помощью вычисления максимального значения метрики f1 на кривой recision_recall_curve для того, чтобы снизить эту ошибку, которая будет снижать потенциальную прибыль.

```python
prob = optuna_search_lgbm_df.predict_proba(features_test)[:, 1]

precisions, recalls, thresholds = precision_recall_curve(target_test, prob)

#Выбор оптимального порога в зависимости от значения f1
f1_scores = 2 * (precisions * recalls) / (precisions + recalls)
best_threshold = thresholds[np.argmax(f1_scores)]

#Применение оптимального порога к предсказаниям
opt_pred = (prob >= best_threshold).astype(int)

#Оценка качества модели при изменённых предсказаниях в соответствии с порогом
precision = precision_score(target_test, opt_pred)
recall = recall_score(target_test, opt_pred)
f1 = f1_score(target_test, opt_pred)

print('Precision:', precision)
print('Recall:', recall)
print('F1-score:', f1)
print('Лучший порог классификации:', best_threshold)

#График precision_recall_curve
plt.figure(figsize=(16, 9))
plt.plot(recalls, precisions, marker='.')
plt.xlabel('Recall', size=18)
plt.ylabel('Precision', size=18)
plt.title('Precision-Recall Кривая', size=20)
plt.grid(True)

# Точка лучшей F1
best_f1_index = np.argmax(f1_scores)
best_precision = precisions[best_f1_index]
best_recall = recalls[best_f1_index]
plt.scatter(best_recall, best_precision, color='red', label='Лучшая F1', s=200)
plt.legend()

# Порог классификации, не уверен что он уместен на графике, но очень хотелось показать его
plt.axvline(x=best_threshold, color='blue', linestyle='--', label='Порог классификации')
plt.legend(fontsize='xx-large')
plt.show()
```

Замечательно, мы нашли порог классификации, при котором достигается максимальная метрика f1, при помощи которой, в свою очередь можно сбалансировать значения precision и recall и улучшить качество модели. Посмотрим как изменилась матрица ошибок.

```python
opt_pred = (prob >= best_threshold).astype(int)

cm = confusion_matrix(target_test, opt_pred)

plt.figure(figsize=(16, 9))
sb.heatmap(cm, annot=True, fmt="d", cmap=sb.color_palette("Blues"), cbar=False)

plt.title('Матрица ошибок', size=20)
plt.xlabel('Предсказанные значения', size=18)
plt.ylabel('Истинные значения', size=18)
plt.show()
```

Количество ложноположительных значений значительно уменьшилось, а количество ложноотрицательных немного увеличилось. Стоит проконсультироваться с отделом финансов и узнать что будет выгодней не применять методы удерживания к клиентам которые на самом деле не собираются уходить или упустить несколько клиентов.
